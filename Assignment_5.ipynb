# To calculate Covariance and Correlation

```Calculate covariance(Age, Strength) and correlation(Age, strength) for following data using```

- Pearson’s Correlation coefficient method
-  Spearman’s rank correlation method

````python
import numpy as np
from scipy.stats import pearsonr, spearmanr
# Provided Data Table
age = np.array([38, 62, 22, 38, 45, 69, 75, 38, 80, 32, 51, 56, 21, 34, 76])
strength = np.array([20, 15, 30, 21, 18, 12, 14, 28, 9, 22, 20, 19, 28, 23, 14])

# First to calculate Pearson's Correlation
pearson_s_correlation = pearsonr(age,strength)
print(pearson_s_correlation)
# Spearman's correlation 
spearman_s_correlation = spearmanr(age,strength)
print(spearman_s_correlation)
# Covariance calculation
covariance_matrix = np.cov(age,strength)
covariance_age_strength = covariance_matrix[0, 1]
print(f"Covariance between Age and Strength: {covariance_age_strength}")
# Correlation Calculation
correlation_matrix = np.corrcoef(age, strength)
correlation_age_strength = correlation_matrix[0, 1]

print(f"Correlation between Age and Strength: {correlation_age_strength}")
````

# To find Correlation and Covaricance using sckit-learn

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from scipy.stats import pearsonr,spearmanr

# Loading the calfornia dataset
housing = fetch_california_housing()

# Create a dataframe from the dataset
data = pd.DataFrame(data=housing.data,columns=housing.feature_names)
data['target']= housing.target

# Calculate Pearson's correlation matrix
pearson_corr_matrix = data.corr(method='pearson')
print(f"The pearson's correlation matrix is : {pearson_corr_matrix}")

# Calculate Spearman's correlation matrix
spearman_corr_matrix = data.corr(method='spearman')
print(f"\n\nThe spearman's correlation matrix is : {spearman_corr_matrix}\n\n")

# plot heatmaps
plt.figure(figsize=(14,10))
plt.subplot(2,2,1)
sns.heatmap(pearson_corr_matrix,annot=True,cmap='coolwarm',fmt='.2f',linewidths=.5)
plt.title("\nPearson's Correlation Heatmap")
plt.subplot(2,2,2)
sns.heatmap(spearman_corr_matrix,annot=True,cmap='coolwarm',fmt='.2f',linewidths=.5)
plt.title("\nSpearman's Correlation Heatmap")
plt.show()
````
