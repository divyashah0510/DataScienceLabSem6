```python
# Step 1: Load data into Spark DataFrame
import pandas as pd
from pyspark.sql import SparkSession

# Initialize Spark session
spark = SparkSession.builder \
    .appName("Exploratory Data Analysis") \
    .getOrCreate()

# Load data into Spark DataFrame
spark_df = spark.read.csv("Data/Mumbai House Prices.csv", header=True, inferSchema=True)

# Step 2: Explore the data with Spark
# Display schema
spark_df.printSchema()

# Display summary statistics
spark_df.describe().show()

# Step 3: Convert to Pandas DataFrame
# Convert Spark DataFrame to Pandas DataFrame
pandas_df = spark_df.toPandas()

# Step 4: EDA with Pandas
# Now you can perform exploratory data analysis using Pandas
# For example:

# Display first few rows
print(pandas_df.head())

# Summary statistics
print(pandas_df.describe())

# Check for missing values
print(pandas_df.isnull().sum())

# Histogram of a numerical column
pandas_df['price_unit'].hist()

# Scatter plot of two numerical columns
import matplotlib.pyplot as plt
plt.scatter(pandas_df['price'], pandas_df['area'])
plt.xlabel('Numerical Column Price')
plt.ylabel('Numerical Column Area')
plt.title('Scatter Plot')
plt.show()

# Close Spark session
spark.stop()
```
